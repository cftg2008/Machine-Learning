## 1. Import libraries

## 2. Read credit card dataset

```python
df = pd.read_csv('/content/drive/My Drive/tech-syntax/ML/Logistic-Regression/creditcard.csv')
df.head()
```

## 3. Describe credit card data elements

* Except the Time and Amount columns, all other columns are normalized. Let's also normalize Time and Amount.

```python
df.describe()
```

## 4. Plot Amount vs Class

* The plot shows most of the fraud transactions were conducted on small amounts.
* Most of the 1s are located on the left side of the graph that indicated smaller transaction amounts.

```python
sns.jointplot(data=df, x=df['Amount'], y=df['Class'])
```

## 5. Normalize the unscaled columns

```python
from sklearn.preprocessing import RobustScaler

# create RobustScaler instance
rs = RobustScaler()

# new dataframe with Time & Amount columns
df_scale = df[['Time','Amount']]

# Use RobustScaler to transform data. StandardScaler is an option but RobustScaler is used because it is less prone to outliers
df_scale = pd.DataFrame(rs.fit_transform(df_scale))

# rename columns in the new dataframe
df_scale.columns = ['scaled_time','scaled_amount']

# join both dataframes before removing duplicate columns
df = pd.concat([df,df_scale], axis=1)

# drop original unscaled columns
df.drop(['Time','Amount'], axis=1, inplace=True)
df.head()
```

## 6. Visualize the non-balanced dataset

```python
# The dataset is very unbalanced because there are only 492 fraudulent transactions or 0.17% of the total volume.
df['Class'].value_counts()
```

```python
# If we use this data to train our model, it will give 99% accuracy because it will just predict non-fraudulent transaction every time. Thus, it is
# not a good model. We need to balance the dataset.

sns.countplot(df['Class'])
```

## 7. Balance the credit card dataset

```python
# filter out all non_fraudulent records
non_fraudulent = df[df['Class'] == 0]

# select only fraudulent transactions
fraudulent = df[df['Class'] == 1]

# use sample method to shuffle records randomly by 100% (frac=1)
non_fraudulent = non_fraudulent.sample(frac=1)

# after shuffling, take out the first 492 non fraudulent records to match the fraudulent ones
non_fraudulent = non_fraudulent[:492]

# balancing the dataset
df2 = pd.concat([non_fraudulent, fraudulent])

# shuffle the combined dataset
df2 = df2.sample(frac=1)
```

## 8. Visualize the balanced dataset

```python
# check the combined dataset for 0s and 1s. They show equal numbers.
df2['Class'].value_counts()
```

```python
# the plot shows both classes
sns.countplot(data=df2, x='Class')
plt.show()
```

## 9. Split the dataset into training and test sets

```python
# Split the dataset by 80% (training) and 20% (testing)
X = df2.drop('Class', axis=1)
y = df2['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

## 10. Train data on different models for comparison

```python
# Logistic Regression Model

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, y_train)

pred = lr.predict(X_test)

# compute metrics :- classification reports, confusion matrices, accuracy scores
print(classification_report(y_test, pred))
print('\n')
print(confusion_matrix(y_test, pred))
print('\n')
print('accuracy is --> ',round(accuracy_score(y_test, pred)*100,2))
```

```python
# Decision Tree Classifier Model

from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)

pred = dtc.predict(X_test)

# compute metrics :- classification reports, confusion matrices, accuracy scores
print(classification_report(y_test, pred))
print('\n')
print(confusion_matrix(y_test, pred))
print('\n')
print('accuracy is --> ',round(accuracy_score(y_test, pred)*100,2))
```

```python
# Random Forest Classifier Model

from sklearn.ensemble import RandomForestClassifier,IsolationForest

rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)

pred = rfc.predict(X_test)

print(classification_report(y_test, pred))
print('\n')
print(confusion_matrix(y_test, pred))
print('\n')
print('accuracy is --> ',round(accuracy_score(y_test, pred)*100,2))
```

## Summary
* We tried the models in Logistic Regression (LR), Decision Tree Classifier (DT), and Random Forest Classifier (RF)
* Both the LR and DT models shows identical accuracy level after several runs
* RF model shows inconsistent accuracy numbers in comparing with the other 2 models. Sometimes the number is higher, while other time the accuracy is lower than there two models. The last run it returned the same accuracy number as LR and DT.
* Overall, it is a good ML practice. Further research is recommended on different models for the best fit.
